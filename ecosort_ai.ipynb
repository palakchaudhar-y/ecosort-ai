```python
from PIL import Image
import os

def clean_folder(folder_path):
    removed = 0
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        try:
            img = Image.open(file_path)
            img.verify()
        except:
            os.remove(file_path)
            removed += 1
    print(f"Removed {removed} bad images from {folder_path}")

clean_folder('dataset/biodegradable')
clean_folder('dataset/non_biodegradable')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=30,
zoom_range=0.3,
horizontal_flip=True,
shear_range=0.2,
brightness_range=[0.7,1.3]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
train_data = train_datagen.flow_from_directory(
    'dataset',
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
val_data = train_datagen.flow_from_directory(
    'dataset',
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(train_data.class_indices)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()

# Block 1
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)))
model.add(MaxPooling2D(2,2))

# Block 2
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

# Block 3
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

# Block 4 (NEW improvement)
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

# Flatten
model.add(Flatten())

# Dense
from tensorflow.keras.regularizers import l2

model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.0005)))
model.add(Dropout(0.4))

# Output
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    callbacks=[early_stop]
)
```

```python
model.save('waste_model.h5')
```

```python
# Upload image
from google.colab import files
uploaded = files.upload()

# Preprocess image
from tensorflow.keras.preprocessing import image
import numpy as np

img_path = list(uploaded.keys())[0]

img = image.load_img(img_path, target_size=(224,224))
img_array = image.img_to_array(img)
img_array = img_array / 255.0
img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)
confidence = prediction[0][0]

print("Confidence:", confidence)

# High confidence cases
if confidence > 0.85:
    print("Non-biodegradable")

elif confidence < 0.15:
    print("Biodegradable")

# 🔥 Uncertain → ask user
else:
    print("Uncertain prediction")
    user_input = input("Describe the waste (food/plastic/metal/etc): ")

    user_input = user_input.lower()

    if any(word in user_input for word in ["food", "fruit", "vegetable", "peel", "organic"]):
        print("Biodegradable (based on user input)")

    elif any(word in user_input for word in ["plastic", "metal", "glass"]):
        print("Non-biodegradable (based on user input)")

    else:
        print("Still unclear — please try another image")
```
